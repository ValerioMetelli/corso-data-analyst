```python
boston_df = pd.read_csv("boston.csv")
boston_df
```

```python
#La media del prezzo delle case cambia a seconda della distanza dal fiume Charles?
print(boston_df.groupby("chas")["medv"].mean().round(2))

print("Come possiamo vedere le case vicino al fiume costano di più")
```

```python
#Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? Come si può spiegare il risultato? 
boston_df["crim"].describe().round(2)
alto_tasso = boston_df["crim"] > 3.61
basso_tasso = boston_df["crim"] < 3.61

print("Prezzo medio con ALTO tasso di criminalità: ", boston_df[alto_tasso]["medv"].mean().round(2))
print("Prezzo medio con BASSO tasso di criminalità: ",boston_df[basso_tasso]["medv"].mean().round(2))
```

```python
#Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? Appare esserci una qualche correlazione? Come si può spiegare il risultato?
mediana_ptratio = boston_df["ptratio"].median()
print("Mediana ptratio:", mediana_ptratio)

alto_ptratio = boston_df["ptratio"] > mediana_ptratio  # Molti alunni per insegnante
basso_ptratio = boston_df["ptratio"] <= mediana_ptratio  # Pochi alunni per insegnante

print("Numero medio stanze con ALTO rapporto alunni-insegnanti:", boston_df[alto_ptratio]["rm"].mean().round(2)) # rapporto ptratio-numero stanze
print("Numero medio stanze con BASSO rapporto alunni-insegnanti:", boston_df[basso_ptratio]["rm"].mean().round(2)) # rapporto ptratio-numero stanze

print("Prezzo medio con ALTO rapporto alunni-insegnanti:", boston_df[alto_ptratio]["medv"].mean().round(2)) # rapporto medv-numero stanze
print("Prezzo medio con BASSO rapporto alunni-insegnanti:", boston_df[basso_ptratio]["medv"].mean().round(2)) # rapporto medv-numero stanze

print("--Possiamo vedere che il rapporto alunni-insegnati non influisce sul numero di stanze medio, ma il costo delle case influisce.")
print("Più il rapporto tra alunni e insegnanti è basso, più le case costano--")
```

```python
#Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni? E del numero di stanze? E delle tasse?
mediana_rad = boston_df["rad"].median() #calcolo la mediana "rad"

alto_rad = boston_df["rad"] > mediana_rad #alto indice accessibilità autostrade
basso_rad = boston_df["rad"] <= mediana_rad #basso indice accessibilità autostrade

print("=== ALTA accessibilità autostrade ===")
print("Età media abitazioni:", boston_df[alto_rad]["age"].mean().round(2))
print("Numero medio stanze:", boston_df[alto_rad]["rm"].mean().round(2))  
print("Tasse medie:", boston_df[alto_rad]["tax"].mean().round(2))

print("\n=== BASSA accessibilità autostrade ===")
print("Età media abitazioni:", boston_df[basso_rad]["age"].mean().round(2))
print("Numero medio stanze:", boston_df[basso_rad]["rm"].mean().round(2))
print("Tasse medie:", boston_df[basso_rad]["tax"].mean().round(2))

print("\nPossiamo vedere come l'accessibilità all'autostrada influisce notevolmente sulle tasse medie e in percentuale più bassa sull'età delle abitazioni")
```

```python
employees_df = pd.DataFrame ({
    'employee_id' : [101, 102, 103, 104, 105],
    'name' : ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'department_id' : [1, 2, 1, 2, 3]
})

departments_df = pd.DataFrame({
    'department_id' : [1, 2, 3],
    'department_name' : ['HR', 'IT', 'Finance'],
    'location' : ['New York', 'San Francisco', 'Chicago']
})

print("EMPLOYEES:")
print(employees_df)
print("\nDEPARTMENTS:")
print(departments_df)
```

```python
#unisco i due df tramite una funzione .merge

merged_df = employees_df.merge(departments_df, on = "department_id", how = 'inner')
merged_df
```

```python
#Setto per entrambi i df 'department_id' come indice, poi unisco i due dataset mediante il metodo .join()
employees_department_id = employees_df.set_index('department_id')
departments_department_id = departments_df.set_index('department_id')

joined_df = employees_department_id.join(departments_department_id, how = 'inner')
joined_df
```

```python
#Ci sono differenze nel risultato? Quali? Perché?
print('Si, la prima è department_id nella join è diventata un indice, mentre nella merge è una colonna.\n Per il resto è tutto invariato')
```

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username") 
password = os.getenv("password") 
host = os.getenv("host") 
dbname = os.getenv("dbname") 

connection_string = "mysql+pymysql://"+username+":"+password+"@"+host+"/"+dbname 

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
# Importo la prima tabella
dimemployee = pd.read_sql("SELECT * FROM dimemployee", db_engine)
dimemployee
```

```python
# Importo la seconda tabella
dimemployeesalesterritory = pd.read_sql("SELECT * FROM dimemployeesalesterritory", db_engine)
dimemployeesalesterritory
```

```python
#effettuo un join tra i due DataFrame usando le colonne EmployeeKey  
first_merge = dimemployee.merge(dimemployeesalesterritory, on='EmployeeKey', how = 'inner')
print(first_merge.shape)

print('\nSi, il risultato è quello atteso.')
```

```python
#importo ora la tabella dimsalesterritory 
dimsalesterritory = pd.read_sql('SELECT * FROM dimsalesterritory', db_engine)
dimsalesterritory
```

```python
#Effettuo un join tra dimsalesterritory e first_merge, usando le colonne SalesTerritoryKey
final_result = first_merge.merge(dimsalesterritory, on = 'SalesTerritoryKey', how = 'inner')
final_result
```

```python
#Conto quanti dipendenti ci sono per ogni paese (country) e per ogni regione (region)
final_result.groupby(['SalesTerritoryCountry', 'SalesTerritoryRegion'])['EmployeeKey'].count()
```

```python
#Valuto la media del BaseRate per ogni paese
final_result.groupby('SalesTerritoryCountry')['BaseRate'].mean().round(2)
```
