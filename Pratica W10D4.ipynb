```python
#Estraiamo la tabella dimproduct dal database AdventureWorks
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username") 
password = os.getenv("password") 
host = os.getenv("host") 
dbname = os.getenv("dbname") 

connection_string = "mysql+pymysql://"+username+":"+password+"@"+host+"/"+dbname 

db_engine = sqlalchemy.create_engine(connection_string)

query = "SELECT * FROM dimproduct"
tabella = pd.read_sql(query, db_engine) 
```

```python
#Seleziono i dati relativi a diverse fasce di età: 20-29, 30-39, 40-50
fascia_20_29 = diabetes_df[(diabetes_df["Age (years)"] >= 20) & (diabetes_df["Age (years)"] <= 29)]
fascia_30_39 = diabetes_df[(diabetes_df["Age (years)"] >= 30) & (diabetes_df["Age (years)"] <= 39)]
fascia_40_50 = diabetes_df[(diabetes_df["Age (years)"] >= 40) & (diabetes_df["Age (years)"] <= 50)]

#Media della pressione sanguigna diastolica per le diverse fasce di età
print("Media fascia 20-29: ", fascia_20_29["Diastolic blood pressure (mm Hg)"].mean().round(2))
print("Media fascia 30-39: ", fascia_30_39["Diastolic blood pressure (mm Hg)"].mean().round(2))
print("Media fascia 40-50: ", fascia_40_50["Diastolic blood pressure (mm Hg)"].mean().round(2))
```

```python
insurance_df = pd.read_csv("insurance.csv")
#Visualizzo le dimensioni, nomi colonne e un'anteprima
print(insurance_df.shape)

print(insurance_df.columns)

insurance_df
```

```python
#Media di charges rispetto a region
insurance_df.groupby("region")["charges"].mean().round(2)
```

```python
#Media di charges rispetto a smoker
insurance_df.groupby("smoker")["charges"].mean().round(2)
```

```python
#Media di charges rispetto a sex
insurance_df.groupby("sex")["charges"].mean().round(2)
```

```python
#Descrittori statistici di bmi 
insurance_df["bmi"].describe().round(2)
```

```python
#minimo di charges rispetto ai quartili di bmi
q1 = insurance_df[insurance_df["bmi"] <= 26.30]
q2 = insurance_df[(insurance_df["bmi"] > 26.30) & (insurance_df["bmi"] <= 30.40)]
q3 = insurance_df[(insurance_df["bmi"] > 30.40) & (insurance_df["bmi"] <= 34.69)]
q4 = insurance_df[insurance_df["bmi"] > 34.69]

print("Primo quartile: ", "Minimo: ", q1["charges"].min())
print("Secondo quartile: ", "Minimo: ", q2["charges"].min())
print("Terzo quartile: ", "Minimo: ", q3["charges"].min())
print("Quarto quartile: ", "Minimo: ", q4["charges"].min())
```

```python
#media di charges rispetto ai quartili di bmi
print("Primo quartile: ", "Media: ", q1["charges"].mean().round(2))
print("Secondo quartile: ", "Media: ", q2["charges"].mean().round(2))
print("Terzo quartile: ", "Media: ", q3["charges"].mean().round(2))
print("Quarto quartile: ", "Media: ", q4["charges"].mean().round(2))
```

```python
#massimo di charges rispetto ai quartili di bmi
print("Primo quartile: ", "Massimo: ", q1["charges"].max())
print("Secondo quartile: ", "Massimo: ", q2["charges"].max())
print("Terzo quartile: ", "Massimo: ", q3["charges"].max())
print("Quarto quartile: ", "Massimo: ", q4["charges"].max())
```
